```python
import matplotlib.pyplot as plt 
import numpy as np
x = np.linspace(-0.3, 2.5) 

y1 = (x ** 2) - 1*x + 2

y2 = y1.copy()
y2[38:42] = 4.2
fig = plt.figure(figsize=(10, 3), facecolor='white')
#fig.set_frameon(False)
plt.plot(x[::-1], y1, '--', color='C0')
plt.plot(x[::-1], y2, '-', lw=3, color='C0')
plt.text(3.5, 3, 'gleam', fontsize=48*2, bbox=dict(fill=False, edgecolor=None, linewidth=0))
plt.xlim(-0.5, 8)
plt.axis("off")
plt.savefig("./logo.png", facecolor='white', transparent=False)
plt.show()
```
